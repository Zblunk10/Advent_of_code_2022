```python
with open("input.txt", "r") as f:
  file_content = f.read().splitlines()
file_content
```

```python
with open("sample.txt", "r") as f:
  sample_content = f.read().splitlines()
sample_content
```

```python
incorrectly_placed_items = []
for line in file_content:
  compartment_size = int(len(line)/2)
  first_compartment =line[:compartment_size]
  second_compartment = line[compartment_size:]
  for item in first_compartment:
    if item in second_compartment:
      incorrectly_placed_items.append(item)
      break


# #check
# print(len(file_content))
# print(len(incorrectly_placed_items))
```

```python
priorities = []
for item in incorrectly_placed_items:
  if item.islower():
    priorities.append(ord(item)-96)
  else:
    priorities.append(ord(item)-38)
```
