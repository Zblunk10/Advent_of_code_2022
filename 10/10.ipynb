### Part 1

```python
with open("input.txt", "r") as f:
  file_content = f.read().splitlines()
```

```python
# with open("sample.txt", "r") as f:
#   file_content = f.read().splitlines()
```

```python
instructions = [x.split() for x in file_content ]
#instructions
```

```python
register_x = 1
cycle_num = 0
results = {}

for line in instructions:
  if line[0] == "noop":
    cycle_num +=1
    results[cycle_num] = register_x

  elif line[0]=="addx":
    cycle_num +=1
    results[cycle_num] = register_x

    cycle_num +=1
    results[cycle_num] = register_x
    register_x += int(line[1])
    

  
```

```python
results
```

```python
#Find the signal strength during the 20th, 60th, 100th, 140th, 180th, and 220th cycles. What is the sum of these six signal strengths?
list_of_wanted = [20,60,100,140,180,220]
wanted_values = []
for i in list_of_wanted:
  wanted_values.append(results[i]*i)
  print(i,results[i], i*results[i] )
print(wanted_values)
print("sum: ", sum(wanted_values))
```

### Part 2

```python
printing = []
for cycle, val_x in results.items():
  position = (cycle-1)%40
#  print("position: ", position, "cycle", cycle, "valx", val_x)
  if position-1 <= val_x <= position+1:
    printing.append("#") 
  else:
    printing.append(".")
```

```python
def insert_newlines(string, every=40):
    return '\n'.join(string[i:i+every] for i in range(0, len(string), every))
new = insert_newlines("".join(printing))
print(new)
```
